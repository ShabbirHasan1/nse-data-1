```python
import pandas as pd
```

```python
df = pd.read_csv('./../lists/zerodha_instrument_list.csv')
```

```python
df
```

```python
dfx = df[df['exchange'].isin(['NSE', 'BSE'])]
```

```python
dfx.head()
```

```python
dfx = dfx.sort_values(by='exchange', ascending=False)
```

```python
dfx = dfx.drop(columns=['last_price', 'strike', 'tick_size', 'lot_size', 'expiry', 'exchange', 'instrument_type'])
```

```python
dfx = dfx.dropna()
```

```python
dfx = dfx.drop_duplicates(subset=['tradingsymbol'], keep='first')
dfx = dfx.drop_duplicates(subset=['name'], keep='first')
```

```python
dfx = dfx.sort_values(by='tradingsymbol')
```

```python
dfx = dfx.reset_index()
```

```python
dfx = dfx.drop(columns=['index'])
```

```python
dfx.where(dfx['tradingsymbol'] == 'TATASTEEL').dropna()
```

```python
dfx.to_csv('./../lists/zerodha_instruments_equities.csv', index=False)
```
